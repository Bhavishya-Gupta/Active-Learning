# Question 1:
## Write a program to load the iris dataset. and perform the following operations:
### 1. Compute histogram of all columns and scatter plots
### 2. Finding the mean, median, mode, and standard deviations of numerical columns

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

iris = pd.read_csv('iris.csv')
sns.set_style("darkgrid")
# Part 1. Compute histogram of all columns and scatter plots
# Histogram
iris.hist(figsize=(13,8),grid=False)
plt.suptitle('Histograms of Iris Dataset Features')
plt.show()

# Scatter plot
sns.pairplot(iris, hue='Species', markers='o')
plt.suptitle('Scatter plots of Iris Dataset Features')
plt.show()

# 2. Finding the mean, median, mode, and standard deviations of numerical columns
numerical_columns = iris.select_dtypes(include=[np.number]).columns

for column in numerical_columns:
    mean_value = iris[column].mean()
    median_value = iris[column].median()
    mode_value = iris[column].mode().iloc[0]  # Handling cases with multiple modes
    std_deviation = iris[column].std()

    print(f"\nStatistics for {column}:")
    print(f"Mean: {mean_value:.3f}")
    print(f"Median: {median_value:.3f}")
    print(f"Mode: {mode_value:.3f}")
    print(f"Standard Deviation: {std_deviation:.3f}")
```
