### Question 3:
#### Using Matplotlib, plot a line graph showing the trend of a numerical dataset. 
#### Customize the plot with title, axis labels, and a legend.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the iris dataset from the CSV file
iris = pd.read_csv('iris.csv')  # Assuming your dataset is named 'iris.csv'

# Plotting the trend lines for each iris species for all four features
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Create subplots with 2 rows and 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

# Increase space between subplots
plt.subplots_adjust(hspace= 1.5, wspace=1.5)

# Flatten the axes for easier indexing
axes = axes.flatten()

# Use a single color palette for all features
color_palette = sns.color_palette('husl', n_colors=len(iris['Species'].unique()))

for i, feature in enumerate(features):
    for j, species in enumerate(iris['Species'].unique()):
        species_data = iris[iris['Species'] == species]
        color = color_palette[j]
        axes[i].plot(species_data[feature], label=species, marker='o', linestyle='-', color=color)

    # Customize each subplot
    axes[i].set_title(f'Trend of {feature} for Iris Species')
    axes[i].set_xlabel('Sample Index')
    axes[i].set_ylabel(feature)
    axes[i].legend()

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

### Question 4:
#### Create a Pandas DataFrame with random values and perform the following operations:
##### - Replace NaN values with the mean of the column.
##### - Sort the DataFrame based on two columns.

```python
import pandas as pd
import numpy as np

# Create a Pandas DataFrame with random values
np.random.seed(42)  # For reproducibility
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10)
}

df = pd.DataFrame(data)
df.iloc[1, 0] = np.nan  # Introduce a NaN value for illustration
df.iloc[4, 2] = np.nan  # Introduce another NaN value for illustration

# Display the original DataFrame
print("Original DataFrame:",df)

# Replace NaN values with the mean of the column
df.fillna(df.mean(), inplace=True)
print("\n")
# Display the DataFrame after replacing NaN values
print("DataFrame after replacing NaN values with column means:",df)

# Sort the DataFrame based on two columns (ascending order)
Sorteddf = df.sort_values(by=['A', 'B'])
print("\n")

# Display the sorted DataFrame
print("DataFrame sorted based on columns 'A' and 'B':",Sorteddf)

```

### Question 5:
#### Using NumPy, calculate the dot product of two matrices and find the determinant of the resulting matrix.

```python
import numpy as np

# Define two matrices
matrix1 = np.array([[1,3,5], [7,9,11]])
matrix2 = np.array([[2,4], [6,8],[10,12]])

# Calculate the dot product of the two matrices
dot_product_result = np.dot(matrix1, matrix2)

# Find the determinant of the resulting matrix
determinant_result = np.linalg.det(dot_product_result)

# Display the original matrices, dot product, and determinant
print("Matrix 1:",matrix1)
print("\n")
print("Matrix 2:",matrix2)
print("\n")
print("Dot Product Result:",dot_product_result)
print("\n")
print("Determinant of the Resulting Matrix:",determinant_result)
```

### Question 6:
#### With Pandas, group a DataFrame by a categorical column and calculate the sum of another numerical column for each group.

```python
import pandas as pd

# Load the iris dataset from the CSV file
iris = pd.read_csv('iris.csv')

# Specify the columns for which you want to calculate the sum
columns_to_sum = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Initialize an empty dictionary to store the results for each column
grouped_data = {}

# Loop through the specified columns
for column in columns_to_sum:
    # Group by the 'species' column and calculate the sum for the current column
    grouped_data[column] = iris.groupby('Species')[column].sum()

# Display the results
print("\n")
print("Sum of selected columns for each 'Species':")
for column, data in grouped_data.items():
    print(f"\nSum of {column} for each 'Species':")
    print(data)
    
```

### Question 9:
#### With Pandas, perform a merge operation between two DataFrames based on a common column.

```python
import pandas as pd

# Load the iris dataset from the CSV file
iris = pd.read_csv('iris.csv')

# Creating two DataFrames from the iris dataset 
df1 = iris[['Species', 'SepalLengthCm', 'SepalWidthCm']]
df2 = iris[['Species', 'PetalLengthCm', 'PetalWidthCm']]

# Display the original DataFrames
print("DataFrame 1:",df1)
print("\n")
print("DataFrame 2:",df2)
# Merge operation based on the common column 'species'
Merged_df = pd.merge(df1, df2, on='Species')

# Display the result of the merge operation
print("Merged DataFrame:",Merged_df)
```

### Question 10:
#### Using Matplotlib, create a scatter plot comparing two numerical variables in a dataset. 
#### Customize the plot by changing the color and shape of the data points.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the iris dataset from the CSV file
iris = pd.read_csv('iris.csv')

# List of numerical columns
numericalcolumns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

# Create scatter plots for each combination in a grid layout
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18,10))

# Flatten the axes for easier indexing
axes = axes.flatten()

# Iterate over combinations and plot
plot_index = 0
for i in range(len(numericalcolumns)):
    for j in range(i + 1, len(numericalcolumns)):
        x_variable = iris[numericalcolumns[i]]
        y_variable = iris[numericalcolumns[j]]

        # Check if there is variability in the data before creating the plot
        if x_variable.nunique() > 1 and y_variable.nunique() > 1:
            # Create scatter plot
            axes[plot_index].scatter(x_variable, y_variable, color="purple", marker="*", label=f'{numericalcolumns[i]} vs {numericalcolumns[j]}')

            # Customize each subplot
            axes[plot_index].set_title(f'Scatter Plot of {numericalcolumns[i]} vs {numericalcolumns[j]}')
            axes[plot_index].set_xlabel(numericalcolumns[i])
            axes[plot_index].set_ylabel(numericalcolumns[j])
            axes[plot_index].legend()

            plot_index += 1

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```
